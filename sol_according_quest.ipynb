# 3.Compute feature reduction based on input. See the screenshot below where there can be No Reduction, Corr with Target, Tree-based, PCA. Please make sure you write code so that all options can work. If we rerun your code with a different Json it should work if we switch No Reduction to say PCA.

```python
#from sklearn.preprocessing import LabelEncoder
def apply_feature_reduction(data: pd.DataFrame, target: pd.Series, reduction_settings: dict):
    method= reduction_settings.get("feature_reduction_method", "No Reduction")
    parameters= reduction_settings.get("parameters", {})
    
    if method == "No Reduction":
        print("No feature reduction applied. Returning all features.")
        return data
    
    elif method == "Corr with Target":
        threshold= parameters.get("threshold", 0.1)
        correlations= data.corrwith(target)
        selected_features= correlations[abs(correlations) >= threshold].index
        print(f"Selected features based on correlation with target: {list(selected_features)}")
        return data[selected_features]
    
    elif method == "Tree-based":
        n_features= parameters.get("n_features", 5)
        model= RandomForestRegressor(random_state=42)
        model.fit(data, target)
        feature_importances= pd.Series(model.feature_importances_, index=data.columns)
        selected_features= feature_importances.nlargest(n_features).index
        print(f"Selected features based on tree-based importance: {list(selected_features)}")
        return data[selected_features]
    
    elif method == "PCA":
        n_components= parameters.get("n_components", 2)
        pca= PCA(n_components= n_components)
        transformed_data= pca.fit_transform(data)
        pca_columns= [f"PCA_{i+1}" for i in range(transformed_data.shape[1])]
        print(f"Applied PCA. Reduced data to {n_components} components.")
        return pd.DataFrame(transformed_data, columns= pca_columns)
    
    else:
        raise ValueError(f"Unsupported feature reduction method: {method}")


def feature_reduction(json_file, csv_file):
    # Extract Feature Handling Information
    feature_reduction_settings= design_state_data["feature_reduction"]
    
    # Encode target variable if categorical
    target= LabelEncoder().fit_transform(df["species"])  # Convert species to integers
    features= df.drop(columns= ["species"])

    # Apply feature reduction
    reduced_data= apply_feature_reduction(features, target, feature_reduction_settings)
    
    print("Reduction completed. Here is the updated DataFrame:")
    print(reduced_data)
    
#Example Usage
feature_reduction("output.json", "iris.csv")
```
